```python
from datetime import datetime
import MetaTrader5 as mt5
import pandas as pd
import pytz
import math as ma
import talib as ta

# display data on the MetaTrader 5 package 
print("MetaTrader5 package author: ",mt5.__author__) 
print("MetaTrader5 package version: ",mt5.__version__) 

# import the 'pandas' module for displaying data obtained in the tabular form
pd.set_option('display.max_columns', 500) # number of columns to be displayed
pd.set_option('display.width', 5500)      # max table width to display

# establish connection to MetaTrader 5 terminal 
if not mt5.initialize():
    print("initialize() failed, error code =",mt5.last_error()) 
    quit()

# set time zone to UTC
timezone = pytz.timezone("Etc/UTC")
#print(timezone)

# create 'datetime' objects in UTC time zone to avoid the implementation of a local time zone offset
utc_from = datetime(2020, 8, 15, tzinfo=timezone)
#utc_to = datetime(2020, 9, 25, hour = 13, tzinfo=timezone)
utc_to = datetime(2020, 9, 15, hour = 23, minute = 59,microsecond = 999999,tzinfo=timezone)
print(utc_from)
print(utc_to)

rates = mt5.copy_rates_range("GOLD", mt5.TIMEFRAME_D1, utc_from, utc_to)
rate1 = mt5.copy_rates_range("GOLD", mt5.TIMEFRAME_H1, utc_from, utc_to)
    
# shut down connection to the MetaTrader 5 terminal
mt5.shutdown() 

# create DataFrame out of the obtained data
rates_frame = pd.DataFrame(rates)
rate1_frame = pd.DataFrame(rate1)

# convert time in seconds into the 'datetime' format
rates_frame['time']=pd.to_datetime(rates_frame['time'], unit='s')
rates_frame['ATR']=round(ta.ATR(rates_frame['high'],rates_frame['low'],rates_frame['close'],timeperiod=1)
,2) 
rate1_frame['time']=pd.to_datetime(rate1_frame['time'], unit='s')
rate1_frame['ATR']=round(ta.ATR(rate1_frame['high'],rate1_frame['low'],rate1_frame['close'],timeperiod=1)
,2) 

print(rates_frame.describe())
print(rate1_frame.describe())

```

```python
rate1_frameAll=rate1_frame.time.count()-1
print(rate1_frameAll)
counter = cAll = 0

pp=cb='y'
cAllTm=0
cAllTp=1
barStatus = rate1_frame.open.loc[rate1_frameAll]-rate1_frame.close.loc[rate1_frameAll]
if barStatus > 0 :
    pp='n'
    cAllTm=1
    cAllTp=0

cm1=cm2=cm3=cm4=cm5=cm6=cm7=cm8=cm9=cm99=0
cp1=cp2=cp3=cp4=cp5=cp6=cp7=cp8=cp9=cp99=0
cAll=1

for rate in range(rate1_frameAll,0,-1):
    counter+=1
    barStatus = rate1_frame.open.loc[rate]-rate1_frame.close.loc[rate]
    #print(barStatus)
    if barStatus > 0:
        #print('-------',rate,pp,cb,rate1_frame.open.loc[rate],rate1_frame.close.loc[rate],barStatus)
        cb = 'y'
        #print('-----222',rate,pp,cb,rate1_frame.open.loc[rate],rate1_frame.close.loc[rate],barStatus)
        if pp == cb :
            cAll += 1
            pp='y'
            #print(cAll)
        else :
            if cAll==1:
                cp1=cp1+cAll
            elif cAll==2:
                cp2=cp2+cAll
            elif cAll==3:
                cp3=cp3+cAll
            elif cAll==4:
                cp4=cp4+cAll
            elif cAll==5:
                cp5=cp5+cAll
            elif cAll==6:
                cp6=cp6+cAll
            elif cAll==7:
                cp7=cp7+cAll
            elif cAll==8:
                cp8=cp8+cAll
            elif cAll==9:
                cp9=cp9+cAll
            else:
                cp99=cp99+cAll
            cAllTp=cAllTp+cAll
            #print('-------------',rate,pp,cb)
            #print(cAll)
            cAll=1
            pp='y'
    else:
        #print('*****')
        #print('-------',rate,pp,cb,rate1_frame.open.loc[rate],rate1_frame.close.loc[rate],barStatus)
        cb = 'n'
        #print('------2-',rate,pp,cb,rate1_frame.open.loc[rate],rate1_frame.close.loc[rate],barStatus)
        if pp == cb :
            #print('//////',rate,pp,cb)
            cAll += 1
            pp='n'
            #print(cAll)
        else :
            if cAll==1:
                cm1=cm1+cAll
            elif cAll==2:
                cm2=cm2+cAll
            elif cAll==3:
                cm3=cm3+cAll
            elif cAll==4:
                cm4=cm4+cAll
            elif cAll==5:
                cm5=cm5+cAll
            elif cAll==6:
                cm6=cm6+cAll
            elif cAll==7:
                cm7=cm7+cAll
            elif cAll==8:
                cm8=cm8+cAll
            elif cAll==9:
                cm9=cm9+cAll
            else:
                cm99=cm99+cAll
            cAllTm=cAllTm+cAll
            #print('----333-',rate,pp,cb,rate1_frame.open.loc[rate],rate1_frame.close.loc[rate],barStatus)
            cAll=1
            pp='n'
print('green= ',cm1,cm2,cm3,cm4,cm5,cm6,cm7,cm8,cm9,cm99,cAllTm)
print('red  = ',cp1,cp2,cp3,cp4,cp5,cp6,cp7,cp8,cp9,cp99,cAllTp)        
    
```
